```python
obs = [[30, 20, 10],
       [20, 25, 15]]
critical = 5.991
```

```python
rows = len(obs)
cols = len(obs[0])

row_sum = []
for row in obs:
    row_sum.append(sum(row))
print(row_sum)

col_sum=[0, 0, 0]
for i in range(rows):
    for j in range(cols):
        col_sum[j]+=obs[i][j]
print(col_sum)

tot = 0
for row in row_sum:
    tot+=row
print(tot)
```

```python
chi = 0
for i in range(rows):
    for j in range(cols):
        e = (row_sum[i]*col_sum[j]) / tot
        o = obs[i][j]
        chi += ((o-e)**2)/e
print(chi)
```

```python
df = (rows-1)*(cols-1)
print(df)
if critical>chi:
    print("Independent")
else:
    print("Dependent")
```
