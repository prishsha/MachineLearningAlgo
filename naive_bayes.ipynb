```python
data = [
    ["Sunny", "High", "Weak", "No"],
    ["Sunny", "High", "Strong", "No"],
    ["Overcast", "High", "Weak", "Yes"],
    ["Rain", "High", "Weak", "Yes"],
    ["Rain", "Normal", "Weak", "Yes"],
    ["Overcast", "Normal", "Strong", "Yes"],
    ["Sunny", "Normal", "Weak", "Yes"],
    ["Rain", "Normal", "Strong", "No"],
    ["Overcast", "High", "Strong", "Yes"],
    ["Sunny", "Normal", "Strong", "Yes"]
]
test = ["Sunny", "High", "Strong"]
```

```python
yes_count = 0
no_count = 0
for row in data:
    if row[3]=="Yes":
        yes_count+=1
    else:
        no_count+=1
print("Yes Count:", yes_count)
print("No Count:", no_count)
```

```python
total_len = len(data)
p_yes = yes_count/ total_len
p_no = no_count/total_len
print("Prior Probability of Yes:", p_yes)
print("Prior Probability of No:", p_no)
```

```python
def cond_prob(index, target, value):
    tot=0
    count=0
    for row in data:
        if row[3]==target:
            tot+=1
            if row[index]==value:
                count+=1
    if tot==0:
        return 0
    return (count/tot)
```

```python
yes = p_yes
no = p_no
for i in range(len(test)):
    yes *= cond_prob(i, "Yes", test[i])
    no *= cond_prob(i, "No", test[i])
print("Prob of Yes:", yes*100)
print("Prob of No:", no*100)
if yes>no:
    print("Final: Yes")
else:
    print("Final: No")
```
