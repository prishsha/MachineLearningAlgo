```python
transactions = [
    ["Laptop", "Mouse", "USB_Drive"],
    ["Smartphone", "Headphones", "Charger"],
    ["Laptop", "Keyboard", "Mouse"],
    ["Tablet", "Stylus", "Case"],
    ["Laptop", "Mouse", "Keyboard", "USB_Drive"],
    ["Smartphone", "Charger", "Screen_Protector"],
    ["Laptop", "Mouse", "USB_Drive", "Keyboard"],
    ["Headphones", "Charger"],
    ["Laptop", "Mouse", "Keyboard"],
    ["Smartphone", "Headphones", "Charger", "Case"],
    ["Tablet", "Stylus", "Case", "Screen_Protector"],
    ["Laptop", "Mouse", "USB_Drive"]
]

min_sup = 4
```

```python
def get_unique_items():
    items=[]
    for row in transactions:
        for item in row:
            if [item] not in items:
                items.append([item])
    return items
```

```python
def get_support(itemset):
    count=0
    for row in transactions:
        if all(item in row for item in itemset):
            count+=1
    return count
```

```python
def generate_candidates(prev, k):
    can=[]
    for i in range(len(prev)):
        for j in range(i+1, len(prev)):
            L1 = prev[i]
            L2 = prev[j]
            L_new = list(set(L1 + L2))
            L_new.sort()
            if len(L_new)==k and L_new not in can:
                can.append(L_new)
    return can
```

```python
def apriori():
    itemset = get_unique_items()

    L1 = []
    for item in itemset:
        if get_support(item) >= min_sup:
            L1.append(item)
    print("L 1:", L1, "\n")
    
    L = L1
    k=2
    while True:
        itemset = generate_candidates(L, k)
        new_L = []
        for item in itemset:
            if get_support(item) >= min_sup:
                new_L.append(item)
        if not new_L:
            break
        print("L", k, ":", new_L, "\n")
        L= new_L
        k+=1

apriori()
```
