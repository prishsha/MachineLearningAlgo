```python
transactions = [
    ["Laptop", "Mouse", "USB_Drive"],
    ["Smartphone", "Headphones", "Charger"],
    ["Laptop", "Keyboard", "Mouse"],
    ["Tablet", "Stylus", "Case"],
    ["Laptop", "Mouse", "Keyboard", "USB_Drive"],
    ["Smartphone", "Charger", "Screen_Protector"],
    ["Laptop", "Mouse", "USB_Drive", "Keyboard"],
    ["Headphones", "Charger"],
    ["Laptop", "Mouse", "Keyboard"],
    ["Smartphone", "Headphones", "Charger", "Case"],
    ["Tablet", "Stylus", "Case", "Screen_Protector"],
    ["Laptop", "Mouse", "USB_Drive"]
]

min_sup = 4
```

```python
def get_unique():
    unq=[]
    for row in transactions:
        for item in row:
            if [item] not in unq:
                unq.append([item])
    return unq
```

```python
def count_support(itemset):
    count = 0
    for row in transactions:
        if all(item in row for item in itemset):
            count+=1
    return count
```

```python
def generate_candidates(prev, k):
    can=[]
    for i in range(len(prev)):
        for j in range(i+1, len(prev)):
            L1 = prev[i]
            L2 = prev[j]
            newL = list(set(L1 + L2))
            newL.sort()
            if len(newL) == k and newL not in can:
                can.append(newL)
    return can
```

```python
def apriori():
    L1 = []
    itemset = get_unique()
    for item in itemset:
        if count_support(item) >= min_sup:
            L1.append(item)
    print("L1:", L1)

    L = L1
    k=2
    while True:
        itemset = generate_candidates(L, k)
        new_L=[]
        for item in itemset:
            if count_support(item) >= min_sup:
                new_L.append(item)
        if not new_L:
            break
        print("L", k, ":", new_L)
        L = new_L
        k+=1
```

```python
apriori()
```
