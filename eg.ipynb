```python
transactions = [
    ["Python", "Java", "C++"],
    ["Python", "R", "SQL"],
    ["Java", "C++", "JavaScript"],
    ["Python", "Java", "R"],
    ["Python", "C++", "SQL"],
    ["Python", "Java", "C++"],
    ["R", "SQL", "MongoDB"]
]
min_sup = 3
```

```python
def generate_frequent():
    freq={}
    for row in transactions:
        for item in row:
            freq[item] = freq.get(item, 0) + 1
    return freq
```

```python
def filter_sort(freq):
    srt=[]
    for row in transactions:
        filtered=[]
        for item in row:
            if freq[item] >=min_sup:
                filtered.append(item)
        filtered.sort(key=lambda x : freq[x], reverse=True)
        if filtered:
            srt.append(filtered)
    return srt
```

```python
class Node:
    def __init__(self, name):
        self.name = name
        self.count = 1
        self.parent = None
        self.children = {}
```

```python
def build_tree(fs):
    root = Node("NULL")
    for row in fs:
        current = root
        for item in row:
            if item in current.children:
                current.children[item].count +=1
            else:
                newn = Node(item)
                current.children[item] = newn
                newn.parent = current
            current = current.children[item]
    return root
```

```python
def print_tree(node, indent=0):
    print(" "*indent, node.name, ":" , node.count)
    for child in node.children.values():
        print_tree(child, indent+4)
```

```python
freq = generate_frequent()
print(freq)
fs = filter_sort(freq)
print(fs)
tree = build_tree(fs)
print_tree(tree)
```
