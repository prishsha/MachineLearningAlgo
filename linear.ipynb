```python
data = [[1,40],
        [2,50],
        [3,60],
        [4,70],
        [5,80]]
test = 6
```

```python
x = []
y = []
n = len(data)
```

```python
for row in data:
    x.append(row[0])
    y.append(row[1])
print(x)
print(y)
print(sum(x))
print(sum(y))
```

```python
x_mean = sum(x)/n
y_mean = sum(y)/n
num=0
den=0
for i in range(n):
    num += (x[i]-x_mean) * (y[i]-y_mean)
    den += (x[i]-x_mean)**2
slope = num/den
print(slope)
```

```python
c = y[0] - slope*x[0]
print(c)
```

```python
pred = slope*test + c
print(pred)
```
